```python
from collections import deque

def bfs_shortest_path(game_data, start_state):
    queue = deque([(start_state, [start_state])])  # (current state, path to current state)
    visited = set()

    while queue:
        current_state, path = queue.popleft()

        if current_state in visited:
            continue

        visited.add(current_state)

        if current_state == "0":
            return path

        for _, info in game_data[current_state]['choices'].items():
            next_state = info[0]
            print(info)
            if next_state not in visited:
                queue.append((str(next_state), path + [str(next_state)]))

    return None  # No path found

# Find the shortest path
start_state = "1"
shortest_path = bfs_shortest_path(gameData, start_state)

if shortest_path:
    print("Shortest path through the game states:", " -> ".join(shortest_path))
else:
    print("No path found from the start state to the terminal state.")
```

```python
def dfs_all_paths(game_data, current_state, path, all_paths):
    if current_state == "0":
        all_paths.add((path, len(path.split(","))))

        return

    seen = set()
    for choice, info in game_data[current_state]['choices'].items():
        next_state = info[0]
        if next_state not in seen:
            seen.add(next_state)
            dfs_all_paths(game_data, str(next_state), path + "," + str(next_state), all_paths)

def find_all_paths(game_data, start_state):
    all_paths = set()
    dfs_all_paths(game_data, start_state, start_state, all_paths)
    return all_paths

# Find all paths
start_state = "1"
all_paths = find_all_paths(gameData, start_state)

# Print all paths and their lengths
for path, length in all_paths:
    print(f"Path: {path}, Length: {length}")
```
